# BINARY SEARCH #
Binary Search is a searching algorithm that works on sorted lists or arrays. It repeatedly divides the search interval in half by comparing the target element with the middle element of the list. If the middle element matches the target element, the search is successful. If the target element is smaller, the search continues in the lower half of the list. If the target element is larger, the search continues in the upper half of the list. This process is repeated until the target element is found or the search interval becomes empty.

The time complexity of binary search is O(log n), where n is the number of elements in the list. This makes it a very efficient searching algorithm for large sorted lists, as the search interval is reduced by half at each step, leading to fast convergence to the target element. However, binary search requires the list to be sorted, and any changes to the list may require re-sorting, which can add additional time complexity. Binary search is commonly used in scenarios where the list is sorted and needs to be searched efficiently, such as in searching for a specific element in a large database or a large collection of data.

```python
import time
import matplotlib.pyplot as plt
```

```python
def binary_search(arr, low, high, x):
    """
    Recursive Binary Search function to search for an element x in a sorted list arr.
    Returns the index of the element if found, otherwise returns -1.
    """
    if low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            return binary_search(arr, mid + 1, high, x)
        else:
            return binary_search(arr, low, mid - 1, x)
    else:
        return -1
```

```python
def time_binary_search(n):
    """
    Function to measure the time taken to perform binary search on a sorted list of n elements.
    Returns the time taken in seconds.
    """
    arr = list(range(n))  # Generate a sorted list of n elements
    x = n - 1  # Element to search for
    start_time = time.time()  # Record start time
    binary_search(arr, 0, len(arr) - 1, x)  # Perform binary search
    end_time = time.time()  # Record end time
    return end_time - start_time
```

```python
# Experiment with different values of n
n_values = list(range(1000, 10001, 1000))  # List of n values from 1000 to 10000 with a step of 1000
time_taken = []  # List to store the time taken for each experiment
```

```python
# Perform the experiments and measure time taken
for n in n_values:
    t = time_binary_search(n)
    time_taken.append(t)
    print(f"Time taken for n = {n}: {t:.6f} seconds")
```

```python

# Plot the graph
plt.plot(n_values, time_taken, marker='o')
plt.xlabel('Number of Elements (n)')
plt.ylabel('Time taken (seconds)')
plt.title('Recursive Binary Search Time Complexity')
plt.grid(True)
plt.show()
```

The above code performs recursive Binary Search on a sorted list of n elements for different values of n, ranging from 1000 to 10000 with a step of 1000. It measures the time taken for each experiment and plots a graph of the time taken versus n using Matplotlib. The graph will show the time complexity of recursive Binary Search, which is expected to be a logarithmic curve, indicating that the time taken increases logarithmically with the number of elements in the sorted list.

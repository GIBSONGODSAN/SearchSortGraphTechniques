# LINEAR SEARCH #
Linear search, also known as sequential search, is a simple searching algorithm that sequentially checks each element in a list or array until the target element is found or the entire list is traversed. It compares the target element with each element in the list one by one, from the beginning to the end, until a match is found or the end of the list is reached.

The time complexity of linear search is O(n), where n is the number of elements in the list. This means that the time taken by linear search increases linearly with the size of the list. It is a straightforward but inefficient searching algorithm for large lists as it may take a long time to search through the entire list, especially if the target element is located towards the end of the list. However, linear search can be useful for small lists or when the list is not sorted, as it does not require any pre-processing or assumptions about the order of elements.

```python
import time
import matplotlib.pyplot as plt
```

```python
def linear_search(arr, x):
    """
    Linear search function to search for an element x in a list arr.
    Returns the index of the element if found, otherwise returns -1.
    """
    for i in range(len(arr)):
        if arr[i] == x:
            return i
    return -1
```

```python
def time_linear_search(n):
    """
    Function to measure the time taken to perform linear search on a list of n elements.
    Returns the time taken in seconds.
    """
    arr = list(range(n))  # Generate a list of n elements
    x = n - 1  # Element to search for
    start_time = time.time()  # Record start time
    linear_search(arr, x)  # Perform linear search
    end_time = time.time()  # Record end time
    return end_time - start_time
```

```python
# Experiment with different values of n
n_values = list(range(1000, 10001, 1000))  # List of n values from 1000 to 10000 with a step of 1000
time_taken = []  # List to store the time taken for each experiment
```

```python
# Perform the experiments and measure time taken
for n in n_values:
    t = time_linear_search(n)
    time_taken.append(t)
    print(f"Time taken for n = {n}: {t:.6f} seconds")
```

```python
# Plot the graph
plt.plot(n_values, time_taken, marker='o')
plt.xlabel('Number of Elements (n)')
plt.ylabel('Time taken (seconds)')
plt.title('Linear Search Time Complexity')
plt.grid(True)
plt.show()
```

The above code performs linear search on a list of n elements for different values of n, ranging from 1000 to 10000 with a step of 1000. It measures the time taken for each experiment and plots a graph of the time taken versus n using Matplotlib. The graph will show the time complexity of linear search, which is expected to be a straight line, indicating that the time taken increases linearly with the number of elements in the list.

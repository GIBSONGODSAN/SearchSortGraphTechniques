# INSERTION SORT
The key idea in Insertion Sort is to partition the list into two parts - the sorted part and the unsorted part. The sorted part starts with only one element (the first element) and grows with each iteration, while the unsorted part gets smaller. The algorithm repeatedly takes an element from the unsorted part (the key) and inserts it into its correct position in the sorted part by comparing and swapping with adjacent elements.

Insertion Sort has an average time complexity of O(n^2), where n is the number of elements in the list. It is not very efficient for large lists, as it requires multiple comparisons and swaps for each element. However, it can be useful for small lists or when the list is partially sorted, as it has a simple implementation and works well in such scenarios.

```python
import time
import matplotlib.pyplot as plt
```

```python
def insertion_sort(arr):
    """
    Insertion Sort function to sort a list of elements in ascending order.
    """
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
```

```python
def time_insertion_sort(n):
    """
    Function to measure the time taken to perform Insertion Sort on a list of n elements.
    Returns the time taken in seconds.
    """
    arr = list(range(n, 0, -1))  # Generate a list of n elements in reverse order
    start_time = time.time()  # Record start time
    insertion_sort(arr)  # Perform Insertion Sort
    end_time = time.time()  # Record end time
    return end_time - start_time
```

```python
# Experiment with different values of n
n_values = list(range(1000, 10001, 1000))  # List of n values from 1000 to 10000 with a step of 1000
time_taken = []  # List to store the time taken for each experiment
```

```python
# Perform the experiments and measure time taken
for n in n_values:
    t = time_insertion_sort(n)
    time_taken.append(t)
    print(f"Time taken for n = {n}: {t:.6f} seconds")
```

```python
# Plot the graph
plt.plot(n_values, time_taken, marker='o')
plt.xlabel('Number of Elements (n)')
plt.ylabel('Time taken (seconds)')
plt.title('Insertion Sort Time Complexity')
plt.grid(True)
plt.show()
```

The above code performs Insertion Sort on a list of n elements for different values of n, ranging from 1000 to 10000 with a step of 1000. It measures the time taken for each experiment and plots a graph of the time taken versus n using Matplotlib. The graph will show the time complexity of Insertion Sort, which is expected to be a quadratic curve, indicating that the time taken increases quadratically with the number of elements in the list. Insertion Sort is not very efficient for large lists, but it can be useful for small lists or when the list is partially sorted, as it has a simple implementation and works well in such scenarios.
